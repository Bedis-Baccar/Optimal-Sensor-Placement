```python
"""
Couverture Connexe Minimum dans les Réseaux de Capteurs
========================================================
"""

import math
import re
import os
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from collections import deque
from typing import Dict, Set, Tuple, List, Optional


# =============================================================================
# PARAMÈTRES GLOBAUX
# =============================================================================
R_CAPT = 1
R_COM = 2
PUITS_X = 0
PUITS_Y = 0


# =============================================================================
# MODULE 1 : PARSERS DE FICHIERS
# =============================================================================

def detecter_format(chemin: str) -> str:
    """Détecte le format du fichier : 'grille_tronquee' ou 'points_aleatoires'."""
    with open(chemin, 'r', encoding='utf-8') as f:
        contenu = f.read().lower()
    
    if 'taille de la grille' in contenu or ':=' in contenu:
        return 'grille_tronquee'
    return 'points_aleatoires'


def parser_grille_tronquee(chemin: str, puits: Tuple[float, float] = (0, 0)) -> Dict[int, Tuple[float, float]]:
    """
    Parse un fichier de grille tronquée.
    Retourne {0: puits, 1: cible1, 2: cible2, ...}
    """
    with open(chemin, 'r', encoding='utf-8') as f:
        contenu = f.read()
    
    # Taille de la grille
    match = re.search(r'taille de la grille compl[èe]te\s*:=\s*(\d+)', contenu, re.IGNORECASE)
    if not match:
        raise ValueError("Taille de grille non trouvée")
    
    taille = int(match.group(1))
    cote = int(math.sqrt(taille))
    
    # Points supprimés
    supprimes = set()
    for m in re.findall(r'\d+\s*:\s*\((\d+),\s*(\d+)\)', contenu):
        supprimes.add((int(m[0]), int(m[1])))
    
    # Construire les points
    points = {0: puits}
    id_point = 1
    for y in range(cote):
        for x in range(cote):
            if (x, y) not in supprimes:
                points[id_point] = (float(x), float(y))
                id_point += 1
    
    return points


def parser_points_aleatoires(chemin: str, puits: Tuple[float, float] = (0, 0)) -> Dict[int, Tuple[float, float]]:
    """
    Parse un fichier de points aléatoires.
    Retourne {0: puits, 1: cible1, 2: cible2, ...}
    """
    with open(chemin, 'r', encoding='utf-8') as f:
        lignes = f.readlines()
    
    points_lus = {}
    for ligne in lignes:
        match = re.match(r'^\s*(\d+)\s+([\d.]+)\s+([\d.]+)', ligne.strip())
        if match:
            id_p = int(match.group(1))
            x, y = float(match.group(2)), float(match.group(3))
            points_lus[id_p] = (x, y)
    
    # Réorganiser avec puits en 0
    if 0 in points_lus:
        # Le point 0 du fichier devient le puits
        return points_lus
    else:
        # Ajouter le puits et décaler les IDs
        points = {0: puits}
        for id_p, coord in points_lus.items():
            points[id_p + 1] = coord
        return points


def charger_instance(chemin: str, puits: Tuple[float, float] = (0, 0)) -> Dict[int, Tuple[float, float]]:
    """
    Charge un fichier et retourne les points.
    Point 0 = puits, Points 1..n = cibles.
    """
    if not os.path.exists(chemin):
        raise FileNotFoundError(f"Fichier non trouvé: {chemin}")
    
    fmt = detecter_format(chemin)
    print(f"Chargement: {os.path.basename(chemin)} (format: {fmt})")
    
    if fmt == 'grille_tronquee':
        points = parser_grille_tronquee(chemin, puits)
    else:
        points = parser_points_aleatoires(chemin, puits)
    
    print(f"  -> {len(points) - 1} cibles + 1 puits")
    return points


def generer_grille(largeur: int, hauteur: int, puits: Tuple[float, float] = (0, 0)) -> Dict[int, Tuple[float, float]]:
    """
    Génère une grille régulière.
    Point 0 = puits, Points 1..n = cibles.
    """
    points = {0: puits}
    id_point = 1
    for y in range(hauteur):
        for x in range(largeur):
            points[id_point] = (float(x), float(y))
            id_point += 1
    
    print(f"Grille {largeur}x{hauteur} générée ({largeur * hauteur} cibles)")
    return points


# =============================================================================
# MODULE 2 : RÉSEAU DE CAPTEURS
# =============================================================================

class ReseauCapteurs:
    """Gère le réseau avec pré-calcul des adjacences."""
    
    def __init__(self, points: Dict[int, Tuple[float, float]], r_capt: float, r_com: float):
        self.points = points
        self.r_capt = r_capt
        self.r_com = r_com
        self.id_puits = 0
        self.cibles = set(points.keys()) - {0}
        
        # Bornes pour affichage
        xs, ys = zip(*points.values())
        self.x_min, self.x_max = min(xs), max(xs)
        self.y_min, self.y_max = min(ys), max(ys)
        
        # Pré-calcul adjacences
        self._calculer_adjacences()
    
    def _distance(self, id1: int, id2: int) -> float:
        x1, y1 = self.points[id1]
        x2, y2 = self.points[id2]
        return math.sqrt((x1 - x2)**2 + (y1 - y2)**2)
    
    def _calculer_adjacences(self):
        """Pré-calcule voisins de captation et communication."""
        self.voisins_capt: Dict[int, List[int]] = {}
        self.voisins_com: Dict[int, List[int]] = {}
        
        tous_ids = list(self.points.keys())
        
        for i in self.cibles:
            # Captation : quelles cibles sont couvertes
            self.voisins_capt[i] = [j for j in self.cibles if self._distance(i, j) <= self.r_capt]
            # Communication : qui peut communiquer avec qui
            self.voisins_com[i] = [j for j in tous_ids if i != j and self._distance(i, j) <= self.r_com]
        
        # Voisins com du puits
        self.voisins_com[0] = [j for j in self.cibles if self._distance(0, j) <= self.r_com]
    
    def cibles_couvertes(self, capteur: int) -> List[int]:
        return self.voisins_capt.get(capteur, [])
    
    def voisins_communication(self, noeud: int) -> List[int]:
        return self.voisins_com.get(noeud, [])


# =============================================================================
# MODULE 3 : HEURISTIQUE GLOUTONNE
# =============================================================================

class HeuristiqueGloutonne:
    """Résout le problème en deux phases."""
    
    def __init__(self, reseau: ReseauCapteurs):
        self.reseau = reseau
    
    def phase1_couverture(self, verbose: bool = True) -> Set[int]:
        """Set Cover glouton."""
        if verbose:
            print("\n[Phase 1] Couverture...")
        
        non_couvertes = self.reseau.cibles.copy()
        selection: Set[int] = set()
        
        while non_couvertes:
            # Trouver le meilleur capteur
            meilleur, score_max = None, 0
            for c in self.reseau.cibles - selection:
                couvertes = set(self.reseau.cibles_couvertes(c))
                score = len(couvertes & non_couvertes)
                if score > score_max:
                    meilleur, score_max = c, score
            
            if meilleur is None:
                break
            
            selection.add(meilleur)
            non_couvertes -= set(self.reseau.cibles_couvertes(meilleur))
            
            if verbose:
                coord = self.reseau.points[meilleur]
                print(f"  + Capteur {meilleur} ({coord[0]:.1f}, {coord[1]:.1f}) : +{score_max} cibles")
        
        if verbose:
            print(f"  => {len(selection)} capteurs pour couverture")
        return selection
    
    def _composante_connexe(self, depart: int, actifs: Set[int]) -> Set[int]:
        """BFS pour trouver la composante connexe."""
        if depart not in actifs:
            return set()
        
        visite = {depart}
        file = deque([depart])
        
        while file:
            courant = file.popleft()
            for v in self.reseau.voisins_communication(courant):
                if v in actifs and v not in visite:
                    visite.add(v)
                    file.append(v)
        
        return visite
    
    def _chemin_vers(self, source: int, destinations: Set[int]) -> Optional[List[int]]:
        """BFS pour trouver le plus court chemin."""
        if source in destinations:
            return []
        
        pred = {source: None}
        file = deque([source])
        arrivee = None
        
        while file and arrivee is None:
            courant = file.popleft()
            for v in self.reseau.voisins_communication(courant):
                if v not in pred:
                    pred[v] = courant
                    if v in destinations:
                        arrivee = v
                        break
                    file.append(v)
        
        if arrivee is None:
            return None
        
        # Reconstruire chemin intermédiaire
        chemin = []
        c = arrivee
        while pred[c] is not None:
            if c != arrivee:
                chemin.append(c)
            c = pred[c]
        return chemin
    
    def phase2_connectivite(self, capteurs: Set[int], verbose: bool = True) -> Set[int]:
        """Ajoute des relais pour connecter au puits."""
        if verbose:
            print("\n[Phase 2] Connectivité...")
        
        resultat = capteurs.copy()
        actifs = resultat | {0}
        
        while True:
            comp_puits = self._composante_connexe(0, actifs)
            isoles = resultat - comp_puits
            
            if not isoles:
                break
            
            if verbose:
                print(f"  {len(isoles)} capteurs isolés")
            
            # Trouver le chemin le plus court
            meilleur_chemin = None
            for iso in isoles:
                chemin = self._chemin_vers(iso, comp_puits)
                if chemin is not None:
                    if meilleur_chemin is None or len(chemin) < len(meilleur_chemin):
                        meilleur_chemin = chemin
            
            if meilleur_chemin is None:
                print("  ERREUR: connexion impossible!")
                break
            
            for relais in meilleur_chemin:
                if relais not in resultat and relais != 0:
                    resultat.add(relais)
                    actifs.add(relais)
                    if verbose:
                        coord = self.reseau.points[relais]
                        print(f"  + Relais {relais} ({coord[0]:.1f}, {coord[1]:.1f})")
        
        nb_relais = len(resultat) - len(capteurs)
        if verbose:
            print(f"  => {nb_relais} relais ajoutés")
        
        return resultat
    
    def resoudre(self, verbose: bool = True) -> Set[int]:
        """Résout le problème complet."""
        capteurs = self.phase1_couverture(verbose)
        solution = self.phase2_connectivite(capteurs, verbose)
        return solution


# =============================================================================
# MODULE 4 : VÉRIFICATION ET VISUALISATION
# =============================================================================

def verifier_solution(solution: Set[int], reseau: ReseauCapteurs) -> bool:
    """Vérifie validité de la solution."""
    print("\n[Vérification]")
    valide = True
    
    # Couverture
    couvertes: Set[int] = set()
    for c in solution:
        couvertes.update(reseau.cibles_couvertes(c))
    
    manquantes = reseau.cibles - couvertes
    if manquantes:
        print(f"  ✗ {len(manquantes)} cibles non couvertes")
        valide = False
    else:
        print(f"  ✓ Toutes les cibles couvertes")
    
    # Connectivité
    actifs = solution | {0}
    visite = {0}
    file = deque([0])
    while file:
        c = file.popleft()
        for v in reseau.voisins_communication(c):
            if v in actifs and v not in visite:
                visite.add(v)
                file.append(v)
    
    non_connectes = solution - visite
    if non_connectes:
        print(f"  ✗ {len(non_connectes)} capteurs non connectés")
        valide = False
    else:
        print(f"  ✓ Tous connectés au puits")
    
    return valide


def visualiser(solution: Set[int], reseau: ReseauCapteurs, titre: str = ""):
    """Affiche le réseau graphiquement."""
    fig, ax = plt.subplots(figsize=(12, 10))
    
    # Séparer les points
    x_puits, y_puits = reseau.points[0]
    x_actifs, y_actifs = [], []
    x_inactifs, y_inactifs = [], []
    
    for i in reseau.cibles:
        x, y = reseau.points[i]
        if i in solution:
            x_actifs.append(x)
            y_actifs.append(y)
        else:
            x_inactifs.append(x)
            y_inactifs.append(y)
    
    # Liens de communication
    noeuds = list(solution) + [0]
    for i, u in enumerate(noeuds):
        for v in noeuds[i+1:]:
            if reseau._distance(u, v) <= reseau.r_com:
                xu, yu = reseau.points[u]
                xv, yv = reseau.points[v]
                ax.plot([xu, xv], [yu, yv], 'g-', alpha=0.3, lw=1)
    
    # Cercles de couverture
    for c in solution:
        x, y = reseau.points[c]
        cercle = patches.Circle((x, y), reseau.r_capt, color='red', alpha=0.1)
        ax.add_patch(cercle)
    
    # Points
    ax.scatter(x_inactifs, y_inactifs, c='lightgrey', s=20, label='Cibles')
    ax.scatter(x_actifs, y_actifs, c='red', s=50, edgecolors='black', label='Capteurs')
    ax.scatter([x_puits], [y_puits], c='black', marker='D', s=100, label='Puits')
    
    # Config
    ax.set_aspect('equal')
    marge = max(reseau.r_capt, reseau.r_com)
    ax.set_xlim(reseau.x_min - marge, reseau.x_max + marge)
    ax.set_ylim(reseau.y_min - marge, reseau.y_max + marge)
    
    titre_final = titre or "Réseau de Capteurs"
    titre_final += f"\n{len(solution)} capteurs (R_capt={reseau.r_capt}, R_com={reseau.r_com})"
    plt.title(titre_final)
    plt.legend(loc='upper right')
    plt.grid(True, ls='--', alpha=0.5)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()


# =============================================================================
# MODULE 5 : FONCTION PRINCIPALE DE RÉSOLUTION
# =============================================================================

def resoudre(points: Dict[int, Tuple[float, float]], 
             r_capt: float = R_CAPT, 
             r_com: float = R_COM,
             verbose: bool = True,
             afficher: bool = True,
             titre: str = "") -> Tuple[Set[int], bool]:
    """
    Résout le problème de couverture connexe minimum.
    
    Args:
        points: Dictionnaire {id: (x, y)} avec id=0 pour le puits
        r_capt: Rayon de captation
        r_com: Rayon de communication
        verbose: Afficher les détails
        afficher: Afficher le graphique
        titre: Titre du graphique
    
    Returns:
        (solution, est_valide): Ensemble des capteurs et validité
    """
    if verbose:
        print("\n" + "="*50)
        print("COUVERTURE CONNEXE MINIMUM")
        print(f"Cibles: {len(points)-1} | R_capt={r_capt} | R_com={r_com}")
        print("="*50)
    
    # Créer le réseau
    reseau = ReseauCapteurs(points, r_capt, r_com)
    
    # Résoudre
    heuristique = HeuristiqueGloutonne(reseau)
    solution = heuristique.resoudre(verbose)
    
    # Vérifier
    valide = verifier_solution(solution, reseau)
    
    if verbose:
        print("\n" + "-"*50)
        print(f"SOLUTION: {len(solution)} capteurs | Valide: {'OUI' if valide else 'NON'}")
        print("-"*50)
    
    # Visualiser
    if afficher:
        visualiser(solution, reseau, titre)
    
    return solution, valide


# =============================================================================
# EXEMPLES D'UTILISATION
# =============================================================================

if __name__ == "__main__":
    
    # -----------------------------------------------------------------
    # EXEMPLE 1 : Grille régulière générée
    # -----------------------------------------------------------------
    print("\n>>> EXEMPLE 1 : Grille 10x10 <<<")
    points_grille = generer_grille(10, 10, puits=(0, 0))
    solution1, valide1 = resoudre(points_grille, r_capt=1, r_com=1, titre="Grille 10x10")
    
    
    # -----------------------------------------------------------------
    # EXEMPLE 2 : Depuis un fichier de grille tronquée
    # -----------------------------------------------------------------
    chemin_grille = r"Projet de métaheuristiques - Instances grilles tronquées\grille1010_1.dat"
    print("\n>>> EXEMPLE 2 : Fichier grille tronquée <<<")
    points_tronquee = charger_instance(chemin_grille)
    solution2, valide2 = resoudre(points_tronquee, r_capt=1, r_com=1, titre="Grille tronquée")
    
    
    # -----------------------------------------------------------------
    # EXEMPLE 3 : Depuis un fichier de points aléatoires
    # -----------------------------------------------------------------
    chemin_aleatoire = r"Projet de métaheuristiques - Instances cibles générées aléatoirement\captANOR900_14_20_2021.dat"
    print("\n>>> EXEMPLE 3 : Fichier points aléatoires <<<")
    points_aleatoires = charger_instance(chemin_aleatoire)
    solution3, valide3 = resoudre(points_aleatoires, r_capt=1, r_com=1, titre="Points aléatoires")
    
    
    # -----------------------------------------------------------------
    # EXEMPLE 4 : Points définis manuellement
    # -----------------------------------------------------------------
    print("\n>>> EXEMPLE 4 : Points manuels <<<")
    points_manuels = {
        0: (0, 0),      # Puits
        1: (1, 0),
        2: (2, 0),
        3: (3, 0),
        4: (1, 1),
        5: (2, 1),
        6: (3, 1),
        7: (2, 2),
        8: (3, 2),
        9: (4, 3),
    }
    solution4, valide4 = resoudre(points_manuels, r_capt=1, r_com=2, titre="Points manuels")
```

```python
# =============================================================================
# MODULE 6 : ALGORITHME GÉNÉTIQUE
# =============================================================================

import random
from typing import Dict, Set, Tuple, List, Optional

class AlgorithmeGenetique:
    """
    Algorithme Génétique Binaire pour le problème de Couverture Connexe Minimum.
    
    Encodage: chromosome[i] = 1 si le capteur d'index i est activé
    Le mapping index <-> ID réel est géré via self.index_to_id et self.id_to_index
    """
    
    def __init__(self, 
                 reseau: ReseauCapteurs,
                 taille_pop: int = 100,
                 nb_generations: int = 200,
                 prob_croisement: float = 0.8,
                 prob_mutation: Optional[float] = None,
                 taille_tournoi: int = 3,
                 penalite: float = 1000.0,
                 elitisme: int = 1,
                 seed: Optional[int] = None):
        """
        Args:
            reseau: Instance de ReseauCapteurs
            taille_pop: Taille de la population
            nb_generations: Nombre de générations
            prob_croisement: Probabilité de croisement
            prob_mutation: Probabilité de mutation par bit (défaut: 1/N)
            taille_tournoi: Taille du tournoi de sélection
            penalite: Poids des pénalités pour contraintes violées
            elitisme: Nombre d'élites à conserver
            seed: Graine aléatoire pour reproductibilité
        """
        self.reseau = reseau
        self.taille_pop = taille_pop
        self.nb_generations = nb_generations
        self.prob_croisement = prob_croisement
        self.taille_tournoi = taille_tournoi
        self.penalite = penalite
        self.elitisme = elitisme
        
        # Mapping index chromosome <-> ID cible réel
        self.index_to_id: List[int] = sorted(list(reseau.cibles))
        self.id_to_index: Dict[int, int] = {id_: idx for idx, id_ in enumerate(self.index_to_id)}
        self.n = len(self.index_to_id)  # Taille du chromosome
        
        # Probabilité de mutation par défaut: 1/N
        self.prob_mutation = prob_mutation if prob_mutation else 1.0 / self.n
        
        # Graine aléatoire
        if seed is not None:
            random.seed(seed)
        
        # Statistiques
        self.historique_fitness: List[float] = []
        self.historique_meilleur: List[float] = []
        self.historique_moyenne: List[float] = []
    
    # -------------------------------------------------------------------------
    # ENCODAGE / DÉCODAGE
    # -------------------------------------------------------------------------
    
    def decoder(self, chromosome: List[int]) -> Set[int]:
        """Convertit un chromosome binaire en ensemble d'IDs de capteurs actifs."""
        return {self.index_to_id[i] for i, gene in enumerate(chromosome) if gene == 1}
    
    def encoder(self, solution: Set[int]) -> List[int]:
        """Convertit un ensemble d'IDs en chromosome binaire."""
        chromosome = [0] * self.n
        for id_capteur in solution:
            if id_capteur in self.id_to_index:
                chromosome[self.id_to_index[id_capteur]] = 1
        return chromosome
    
    # -------------------------------------------------------------------------
    # FONCTION DE FITNESS
    # -------------------------------------------------------------------------
    
    def calculer_fitness(self, chromosome: List[int]) -> float:
        """
        Calcule la fitness (à minimiser).
        Fitness = nb_capteurs + pénalité_couverture + pénalité_connectivité
        """
        capteurs_actifs = self.decoder(chromosome)
        nb_capteurs = len(capteurs_actifs)
        
        if nb_capteurs == 0:
            # Aucun capteur: pénalité maximale
            return self.penalite * len(self.reseau.cibles) * 2
        
        # --- Pénalité de couverture ---
        cibles_couvertes: Set[int] = set()
        for c in capteurs_actifs:
            cibles_couvertes.update(self.reseau.cibles_couvertes(c))
        
        nb_non_couvertes = len(self.reseau.cibles - cibles_couvertes)
        penalite_couverture = self.penalite * nb_non_couvertes
        
        # --- Pénalité de connectivité (BFS depuis le puits) ---
        actifs_avec_puits = capteurs_actifs | {0}
        connectes = self._bfs_connectes(0, actifs_avec_puits)
        nb_non_connectes = len(capteurs_actifs - connectes)
        penalite_connectivite = self.penalite * nb_non_connectes
        
        return nb_capteurs + penalite_couverture + penalite_connectivite
    
    def _bfs_connectes(self, depart: int, actifs: Set[int]) -> Set[int]:
        """BFS pour trouver tous les nœuds connectés au départ."""
        visite = {depart}
        file = deque([depart])
        
        while file:
            courant = file.popleft()
            for voisin in self.reseau.voisins_communication(courant):
                if voisin in actifs and voisin not in visite:
                    visite.add(voisin)
                    file.append(voisin)
        
        return visite - {0}  # Retourne seulement les capteurs (pas le puits)
    
    # -------------------------------------------------------------------------
    # INITIALISATION DE LA POPULATION
    # -------------------------------------------------------------------------
    
    def generer_individu_aleatoire(self) -> List[int]:
        """Génère un chromosome aléatoire."""
        # Probabilité d'activation basée sur une estimation du ratio optimal
        prob_activation = min(0.5, 10.0 / self.n + 0.1)
        return [1 if random.random() < prob_activation else 0 for _ in range(self.n)]
    
    def initialiser_population(self, solution_initiale: Optional[Set[int]] = None) -> List[List[int]]:
        """
        Initialise la population.
        Si une solution initiale est fournie, elle est injectée.
        """
        population = []
        
        # Injecter la solution gloutonne si fournie
        if solution_initiale:
            chromosome_init = self.encoder(solution_initiale)
            population.append(chromosome_init)
            # Ajouter quelques variations de la solution initiale
            for _ in range(min(5, self.taille_pop // 10)):
                mutant = chromosome_init.copy()
                # Mutation légère
                for i in range(self.n):
                    if random.random() < self.prob_mutation * 2:
                        mutant[i] = 1 - mutant[i]
                population.append(mutant)
        
        # Compléter avec des individus aléatoires
        while len(population) < self.taille_pop:
            population.append(self.generer_individu_aleatoire())
        
        return population
    
    # -------------------------------------------------------------------------
    # OPÉRATEURS GÉNÉTIQUES
    # -------------------------------------------------------------------------
    
    def selection_tournoi(self, population: List[List[int]], 
                          fitness_pop: List[float]) -> List[int]:
        """Sélection par tournoi."""
        indices = random.sample(range(len(population)), self.taille_tournoi)
        meilleur_idx = min(indices, key=lambda i: fitness_pop[i])
        return population[meilleur_idx].copy()
    
    def croisement_uniforme(self, parent1: List[int], parent2: List[int]) -> Tuple[List[int], List[int]]:
        """Croisement uniforme."""
        enfant1, enfant2 = [], []
        for g1, g2 in zip(parent1, parent2):
            if random.random() < 0.5:
                enfant1.append(g1)
                enfant2.append(g2)
            else:
                enfant1.append(g2)
                enfant2.append(g1)
        return enfant1, enfant2
    
    def croisement_un_point(self, parent1: List[int], parent2: List[int]) -> Tuple[List[int], List[int]]:
        """Croisement en un point."""
        point = random.randint(1, self.n - 1)
        enfant1 = parent1[:point] + parent2[point:]
        enfant2 = parent2[:point] + parent1[point:]
        return enfant1, enfant2
    
    def mutation(self, chromosome: List[int]) -> List[int]:
        """Mutation bit-flip."""
        mutant = chromosome.copy()
        for i in range(self.n):
            if random.random() < self.prob_mutation:
                mutant[i] = 1 - mutant[i]
        return mutant
    
    # -------------------------------------------------------------------------
    # BOUCLE PRINCIPALE
    # -------------------------------------------------------------------------
    
    def run(self, 
            solution_initiale: Optional[Set[int]] = None,
            verbose: bool = True,
            afficher_progression: bool = True) -> Set[int]:
        """
        Exécute l'algorithme génétique.
        
        Args:
            solution_initiale: Solution à injecter (ex: solution gloutonne)
            verbose: Afficher les détails
            afficher_progression: Afficher la courbe de convergence
        
        Returns:
            Meilleure solution trouvée (Set d'IDs de capteurs)
        """
        if verbose:
            print("\n" + "="*50)
            print("ALGORITHME GÉNÉTIQUE")
            print(f"Population: {self.taille_pop} | Générations: {self.nb_generations}")
            print(f"Croisement: {self.prob_croisement:.0%} | Mutation: {self.prob_mutation:.4f}")
            print("="*50)
        
        # Initialisation
        population = self.initialiser_population(solution_initiale)
        fitness_pop = [self.calculer_fitness(ind) for ind in population]
        
        # Meilleur global
        idx_meilleur = min(range(len(population)), key=lambda i: fitness_pop[i])
        meilleur_global = population[idx_meilleur].copy()
        fitness_meilleur_global = fitness_pop[idx_meilleur]
        
        if verbose and solution_initiale:
            fitness_init = self.calculer_fitness(self.encoder(solution_initiale))
            print(f"Fitness solution initiale: {fitness_init:.1f}")
        
        # Boucle évolutive
        for gen in range(self.nb_generations):
            nouvelle_pop = []
            
            # Élitisme
            indices_tries = sorted(range(len(population)), key=lambda i: fitness_pop[i])
            for i in range(self.elitisme):
                nouvelle_pop.append(population[indices_tries[i]].copy())
            
            # Génération des enfants
            while len(nouvelle_pop) < self.taille_pop:
                # Sélection
                parent1 = self.selection_tournoi(population, fitness_pop)
                parent2 = self.selection_tournoi(population, fitness_pop)
                
                # Croisement
                if random.random() < self.prob_croisement:
                    enfant1, enfant2 = self.croisement_uniforme(parent1, parent2)
                else:
                    enfant1, enfant2 = parent1.copy(), parent2.copy()
                
                # Mutation
                enfant1 = self.mutation(enfant1)
                enfant2 = self.mutation(enfant2)
                
                nouvelle_pop.append(enfant1)
                if len(nouvelle_pop) < self.taille_pop:
                    nouvelle_pop.append(enfant2)
            
            # Évaluation
            population = nouvelle_pop
            fitness_pop = [self.calculer_fitness(ind) for ind in population]
            
            # Mise à jour meilleur global
            idx_meilleur_gen = min(range(len(population)), key=lambda i: fitness_pop[i])
            if fitness_pop[idx_meilleur_gen] < fitness_meilleur_global:
                meilleur_global = population[idx_meilleur_gen].copy()
                fitness_meilleur_global = fitness_pop[idx_meilleur_gen]
            
            # Statistiques
            self.historique_meilleur.append(fitness_meilleur_global)
            self.historique_moyenne.append(sum(fitness_pop) / len(fitness_pop))
            
            # Affichage périodique
            if verbose and (gen + 1) % 20 == 0:
                print(f"  Génération {gen+1:4d} | Meilleur: {fitness_meilleur_global:6.1f} | "
                      f"Moyenne: {self.historique_moyenne[-1]:6.1f}")
        
        # Résultat final
        solution_finale = self.decoder(meilleur_global)
        
        if verbose:
            print("-"*50)
            print(f"RÉSULTAT AG: {len(solution_finale)} capteurs (fitness: {fitness_meilleur_global:.1f})")
        
        # Courbe de convergence
        if afficher_progression:
            self.afficher_convergence()
        
        return solution_finale
    
    def afficher_convergence(self):
        """Affiche la courbe de convergence."""
        plt.figure(figsize=(10, 5))
        generations = range(1, len(self.historique_meilleur) + 1)
        
        plt.plot(generations, self.historique_meilleur, 'b-', label='Meilleur', linewidth=2)
        plt.plot(generations, self.historique_moyenne, 'r--', label='Moyenne', alpha=0.7)
        
        plt.xlabel('Génération')
        plt.ylabel('Fitness')
        plt.title('Convergence de l\'Algorithme Génétique')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()


# =============================================================================
# FONCTION DE RÉSOLUTION AVEC AG
# =============================================================================

def resoudre_ag(points: Dict[int, Tuple[float, float]], 
                r_capt: float = R_CAPT, 
                r_com: float = R_COM,
                taille_pop: int = 100,
                nb_generations: int = 200,
                utiliser_glouton: bool = True,
                verbose: bool = True,
                afficher: bool = True,
                titre: str = "") -> Tuple[Set[int], bool]:
    """
    Résout le problème avec l'Algorithme Génétique.
    
    Args:
        points: Dictionnaire {id: (x, y)} avec id=0 pour le puits
        r_capt: Rayon de captation
        r_com: Rayon de communication
        taille_pop: Taille de la population
        nb_generations: Nombre de générations
        utiliser_glouton: Injecter la solution gloutonne dans la population initiale
        verbose: Afficher les détails
        afficher: Afficher le graphique
        titre: Titre du graphique
    
    Returns:
        (solution, est_valide)
    """
    # Créer le réseau
    reseau = ReseauCapteurs(points, r_capt, r_com)
    
    # Solution gloutonne pour initialisation
    solution_glouton = None
    if utiliser_glouton:
        heuristique = HeuristiqueGloutonne(reseau)
        solution_glouton = heuristique.resoudre(verbose=False)
        if verbose:
            print(f"Solution gloutonne: {len(solution_glouton)} capteurs")
    
    # Algorithme Génétique
    ag = AlgorithmeGenetique(
        reseau=reseau,
        taille_pop=taille_pop,
        nb_generations=nb_generations
    )
    solution = ag.run(solution_initiale=solution_glouton, verbose=verbose)
    
    # Vérification
    valide = verifier_solution(solution, reseau)
    
    # Visualisation
    if afficher:
        visualiser(solution, reseau, titre or "Solution AG")
    
    return solution, valide


# =============================================================================
# COMPARAISON GLOUTON vs AG
# =============================================================================

def comparer_methodes(points: Dict[int, Tuple[float, float]],
                      r_capt: float = R_CAPT,
                      r_com: float = R_COM,
                      nb_generations: int = 200,
                      afficher: bool = True) -> Dict:
    """
    Compare l'heuristique gloutonne et l'algorithme génétique.
    
    Returns:
        Dictionnaire avec les résultats comparatifs
    """
    import time
    
    print("\n" + "="*60)
    print("COMPARAISON: GLOUTON vs ALGORITHME GÉNÉTIQUE")
    print(f"Cibles: {len(points)-1} | R_capt={r_capt} | R_com={r_com}")
    print("="*60)
    
    reseau = ReseauCapteurs(points, r_capt, r_com)
    
    # --- GLOUTON ---
    print("\n>>> HEURISTIQUE GLOUTONNE <<<")
    t0 = time.time()
    heuristique = HeuristiqueGloutonne(reseau)
    solution_glouton = heuristique.resoudre(verbose=False)
    temps_glouton = time.time() - t0
    valide_glouton = verifier_solution(solution_glouton, reseau)
    print(f"Temps: {temps_glouton:.3f}s | Capteurs: {len(solution_glouton)}")
    
    # --- AG ---
    print("\n>>> ALGORITHME GÉNÉTIQUE <<<")
    t0 = time.time()
    ag = AlgorithmeGenetique(reseau, nb_generations=nb_generations)
    solution_ag = ag.run(solution_initiale=solution_glouton, 
                         verbose=True, 
                         afficher_progression=afficher)
    temps_ag = time.time() - t0
    valide_ag = verifier_solution(solution_ag, reseau)
    print(f"Temps: {temps_ag:.3f}s | Capteurs: {len(solution_ag)}")
    
    # --- RÉSUMÉ ---
    print("\n" + "-"*60)
    print("RÉSUMÉ COMPARATIF")
    print("-"*60)
    print(f"{'Méthode':<20} {'Capteurs':<12} {'Valide':<10} {'Temps':<10}")
    print(f"{'Glouton':<20} {len(solution_glouton):<12} {'OUI' if valide_glouton else 'NON':<10} {temps_glouton:.3f}s")
    print(f"{'AG':<20} {len(solution_ag):<12} {'OUI' if valide_ag else 'NON':<10} {temps_ag:.3f}s")
    
    amelioration = len(solution_glouton) - len(solution_ag)
    if amelioration > 0:
        print(f"\n✓ AG améliore de {amelioration} capteurs ({amelioration/len(solution_glouton)*100:.1f}%)")
    elif amelioration < 0:
        print(f"\n✗ Glouton meilleur de {-amelioration} capteurs")
    else:
        print(f"\n= Résultats identiques")
    
    # Visualisation côte à côte
    if afficher:
        fig, axes = plt.subplots(1, 2, figsize=(16, 7))
        
        for ax, sol, titre in [(axes[0], solution_glouton, f"Glouton ({len(solution_glouton)} capteurs)"),
                                (axes[1], solution_ag, f"AG ({len(solution_ag)} capteurs)")]:
            # Points
            x_puits, y_puits = reseau.points[0]
            x_actifs = [reseau.points[i][0] for i in sol]
            y_actifs = [reseau.points[i][1] for i in sol]
            x_inactifs = [reseau.points[i][0] for i in reseau.cibles - sol]
            y_inactifs = [reseau.points[i][1] for i in reseau.cibles - sol]
            
            # Cercles de couverture
            for c in sol:
                x, y = reseau.points[c]
                cercle = patches.Circle((x, y), r_capt, color='red', alpha=0.1)
                ax.add_patch(cercle)
            
            ax.scatter(x_inactifs, y_inactifs, c='lightgrey', s=15)
            ax.scatter(x_actifs, y_actifs, c='red', s=40, edgecolors='black')
            ax.scatter([x_puits], [y_puits], c='black', marker='D', s=80)
            
            ax.set_aspect('equal')
            ax.set_title(titre)
            ax.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
    
    return {
        'glouton': {'solution': solution_glouton, 'valide': valide_glouton, 'temps': temps_glouton},
        'ag': {'solution': solution_ag, 'valide': valide_ag, 'temps': temps_ag},
        'amelioration': amelioration
    }
# =============================================================================
# TESTS COMPARATIFS
# =============================================================================

if __name__ == "__main__":
    
    # -----------------------------------------------------------------
    # TEST 1 : Grille régulière
    # -----------------------------------------------------------------
    print("\n" + "#"*60)
    print("# TEST 1 : Grille 10x10")
    print("#"*60)
    
    points_grille = generer_grille(10, 10, puits=(0, 0))
    resultats1 = comparer_methodes(points_grille, r_capt=1, r_com=1, nb_generations=150)
    
    
    # -----------------------------------------------------------------
    # TEST 2 : Grille tronquée (décommenter si fichier disponible)
    # -----------------------------------------------------------------
    print("\n" + "#"*60)
    print("# TEST 2 : Grille tronquée")
    print("#"*60)
    # 
    chemin_grille = r"Projet de métaheuristiques - Instances grilles tronquées\grille1010_1.dat"
    points_tronquee = charger_instance(chemin_grille)
    resultats2 = comparer_methodes(points_tronquee, r_capt=1, r_com=1, nb_generations=150)
    
    
    # -----------------------------------------------------------------
    # TEST 3 : Points aléatoires (décommenter si fichier disponible)
    # -----------------------------------------------------------------
    print("\n" + "#"*60)
    print("# TEST 3 : Points aléatoires")
    print("#"*60)
    # 
    chemin_aleatoire = r"Projet de métaheuristiques - Instances cibles générées aléatoirement\captANOR900_14_20_2021.dat"
    points_aleatoires = charger_instance(chemin_aleatoire)
    resultats3 = comparer_methodes(points_aleatoires, r_capt=1, r_com=1, nb_generations=200)
    
    
    # -----------------------------------------------------------------
    # TEST 4 : AG seul avec paramètres personnalisés
    # -----------------------------------------------------------------
    print("\n" + "#"*60)
    print("# TEST 4 : AG avec paramètres avancés")
    print("#"*60)
    
    points_test = generer_grille(8, 8, puits=(0, 0))
    reseau_test = ReseauCapteurs(points_test, r_capt=1, r_com=1)
    
    ag_custom = AlgorithmeGenetique(
        reseau=reseau_test,
        taille_pop=150,
        nb_generations=100,
        prob_croisement=0.85,
        prob_mutation=0.02,
        taille_tournoi=4,
        elitisme=2
    )
    
    # Sans solution initiale
    solution_ag_seul = ag_custom.run(solution_initiale=None, verbose=True)
    verifier_solution(solution_ag_seul, reseau_test)
    visualiser(solution_ag_seul, reseau_test, "AG seul (sans glouton)")
```
